```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('LatLonFiltered.csv')
df['Fecha'] = pd.to_datetime(df['Fecha'])
                                                                                                                             # Crear columnas retrasadas
for i in range(1, 13):  # Crear 3 columnas retrasadas como ejemplo
    df[f'SST_Lag{i}'] = df['SST'].shift(i)

# Eliminar filas con valores NaN creados por el desplazamiento
df.dropna(inplace=True)

# Configuración de la ventana deslizante
window_size = 12
prediction_step = 1
train_start = 0
train_end = train_start + window_size

# Predicciones
predictions = []
actual = []
```

```python
while train_end + prediction_step - 1 < len(df):
    # Datos de entrenamiento y prueba
    train_data = df.iloc[train_start:train_end]
    test_data = df.iloc[train_end + prediction_step - 1:train_end + prediction_step]

    # Asegurarse de que estamos en el rango de fechas deseado (2020-2023)
    if test_data['Fecha'].dt.year.values[0] < 2020:
        train_start += 1
        train_end += 1
        continue
    if test_data['Fecha'].dt.year.values[0] > 2022:
        break

    # Características y variable objetivo
    features = [f'SST_Lag{i}' for i in range(1, 13)]
    X_train = train_data[features]
    y_train = train_data['SST']
    X_test = test_data[features]

    # Entrenamiento del modelo
    model = RandomForestRegressor(n_estimators=50)
    model.fit(X_train, y_train)

    # Predicción
    pred = model.predict(X_test)
    predictions.append(pred[0])
    actual.append(test_data['SST'].values[0])

    # Mover la ventana
    train_start += 1
    train_end += 1
```

```python

resultados = pd.DataFrame({
    'Prediccion_SST': predictions
})
        
```

```python
resultados.to_csv("output_rndForest1.csv", index=False)

df_adjust = df[(df["Fecha"] >= '2020-01-01') & (df['Fecha'] < '2023-01-01')]

df_adjust = df_adjust.reset_index(drop=True)

df_adjust = pd.DataFrame(df_adjust)
```

```python
mse = mean_squared_error(actual, predictions)
print('Mean Squared Error:', mse)
```
